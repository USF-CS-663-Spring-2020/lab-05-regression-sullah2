```python
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("datasets/Credit.csv")
```

```python
df
```

```python
df = df.iloc[:,1:]
```

```python
X = df[['Income']].values.reshape(-1,1)
y = df[['Limit']].values.reshape(-1,1)
```

```python
X, y
```

```python
plt.scatter(x=X, y = y, color = 'black')
plt.xlabel('Income')
plt.ylabel('Limit')
```

```python
r1 = sns.residplot(X, y)
r1.set(xlabel = 'Income', ylabel = 'Residual')
```

```python
reg_ = LinearRegression()
```

```python
reg_.fit(X, y)
print(reg_.coef_)
print(reg_.intercept_)
```
