```python
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("datasets/toluca.csv")
```

```python
df.head()
```

```python
X = df['workHours'].values.reshape(-1,1)
y = df['lotSize'].values.reshape(-1,1)
```

```python
plt.scatter(x=X, y = y, color = 'black')
plt.xlabel('Work Hours')
plt.ylabel('LotSize')
```

```python
r1 = sns.residplot(y, X)
r1.set(xlabel = 'Lot Size', ylabel = 'Residual')
```

```python
r1 = sns.residplot(X, y)
r1.set(xlabel = 'Work Hours', ylabel = 'Residual')
```

```python
reg_ = LinearRegression()
```

```python
reg_.fit(X, y)
```

```python
reg_.coef_
```

```python
reg_.intercept_
```
